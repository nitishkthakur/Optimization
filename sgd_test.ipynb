```python
from sgd import sgd_base
%load_ext autoreload
%autoreload 1
%aimport sgd.sgd_base

import numpy as np
import matplotlib.pyplot as plt

```

```python
def squared_error_loss(x):
    """
    Compute the squared error loss between true and predicted values.

    Returns:
    float: Squared error loss.
    """
    return (x) ** 2


x = np.arange(-10, 10, 0.1)
y = squared_error_loss(x)

# Plotting the squared error loss

plt.plot(x, y)
plt.title('Squared Error Loss')
plt.xlabel('x')
plt.ylabel('Loss')
plt.grid()
plt.show()
```

```python
optimizer = sgd_base.SGDBase(learning_rate=0.1)

# Example usage of the optimizer
params = np.array([100.0])

# Optimize
for _ in range(100):
    optimizer.step(params=params, loss=squared_error_loss)
    params = optimizer.params

# Print the optimized parameters
optimizer.plot_loss_vs_iterations()
```
